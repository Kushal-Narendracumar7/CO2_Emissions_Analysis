```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv('Co2_Emissions_by_Sectors.csv')
```

```python
df.isnull().sum()
```

```python
total = df['Agriculture_Co2_Emissions_MetricTons'] + df['Domestic_Co2_Emissions_MetricTons'] + df['Automobile_Co2_Emissions_MetricTons'] + df['Industrial_Co2_Emissions_MetricTons']
```

```python
total
```

```python
df['Co2_Emissions_MetricTons']
```

```python
df['Sector_Total'] = df['Agriculture_Co2_Emissions_MetricTons'] + \
                     df['Domestic_Co2_Emissions_MetricTons'] + \
                     df['Automobile_Co2_Emissions_MetricTons'] + \
                     df['Industrial_Co2_Emissions_MetricTons']

df['Difference'] = df['Sector_Total'] - df['Co2_Emissions_MetricTons']
```

```python
df['Difference']
```

```python
px.histogram(
    df,
    x='Difference',
    title='Difference between Sector Total and Co2 Emissions',

).show()
```
